```python
import numpy as np
import matplotlib.pyplot as plt

def generate_data(n,gtype):
    ϵ_values=[]
    for i in range(n):
        if gtype=='u':
            e=np.random.uniform(0,1)
        else:
            e=np.random.randn()
        ϵ_values.append(e)
    return ϵ_values

x=generate_data(100,'u')
plt.plot(x)
plt.show()
```

```python
#list comprehension syntax
import numpy as np
import matplotlib.pyplot as plt

def generate_data(n,gtype):
    ϵ_values=[gtype() for i in range(n)]
    return ϵ_values

x=generate_data(100,np.random.uniform)
plt.plot(x)
plt.show()
```

```python
#EX1
def factorial(n):
    result=1
    for i in range(1,n+1):
        result=i*result
    return result

x=factorial(4)
print(x)
```

```python
#EX2
import numpy as np
def binomial_rv(n,p):
    count=0
    for i in range(n):
        u=np.random.uniform()
        if u<p:
            count+=1
    return count

binomial_rv(10,0.5)
```

```python
#EX3
import numpy as np
c=0
n=10000
for i in range(n):
    x=np.random.uniform(-1,1)
    y=np.random.uniform(-1,1)
    if x**2 + y**2<=1:
        c=c+1
print(4*(c/n))
        
```

```python
#EX4
from numpy.random import uniform
def coinflip(n,p):
    count=0
    u=[]
    for i in range(n):
        e=uniform(0,1)
        u.append(e)
    for i in range(n-2):
        if u[i]<p and u[i+1]<p and u[i+2]<p:
            count+=1
    return count

coinflip(10,0.5)
    
        
```

```python
#EX5(see solution below is more simple)
import numpy as np
import matplotlib.pyplot as plt
def corr_time_s(t,α):
    xx=0
    x=[]
    for i in range(t):
        e=np.random.randn()
        xx=α*xx+e
        x.append(xx)
    return x


y=corr_time_s(200,0.9)
plt.plot(y)
plt.show()

```

```python
#EX6
import numpy as np
import matplotlib.pyplot as plt
def corr_time_s(t,α):
    xx=0
    x=[]
    for i in range(t):
        e=np.random.randn()
        xx=α*xx+e
        x.append(xx)
    return x

A=[0,0.8,0.98]
y=[]
for α in A:
    yy=corr_time_s(200,α)
    y.append(yy)
    
plt.plot(y[0],label='α=0')
plt.plot(y[1],label='α=0.8')
plt.plot(y[2],label='α=0.98')
plt.legend()
plt.show()
    
```
